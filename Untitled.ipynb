```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt


# ---- load image ----
def load_image(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"Image not found: {path}")

    image = cv2.imread(path)
    if image is None:
        raise ValueError("Could not read the image.")

    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


# ---- grayscale ----
def convert_to_grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)


# ---- custom otsu (manual implementation) ----
def custom_otsu(gray):
    hist = np.histogram(gray.flatten(), bins=256)[0]
    total = gray.size

    sum_total = 0
    for i in range(256):
        sum_total += i * hist[i]

    sumB = 0
    wB = 0
    max_var = 0
    threshold = 0

    for t in range(256):
        wB += hist[t]
        if wB == 0:
            continue

        wF = total - wB
        if wF == 0:
            break

        sumB += t * hist[t]

        mB = sumB / wB
        mF = (sum_total - sumB) / wF

        var_between = wB * wF * (mB - mF) ** 2

        if var_between > max_var:
            max_var = var_between
            threshold = t

    _, mask = cv2.threshold(gray, threshold, 255, cv2.THRESH_BINARY)
    return threshold, mask


# ---- morphology ----
def apply_morphology(mask, kernel_size=3):
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    opened = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    cleaned = cv2.morphologyEx(opened, cv2.MORPH_CLOSE, kernel)
    return cleaned


# ---- display ----
def show_results(image, gray, mask_raw, mask_cleaned, threshold):
    plt.figure(figsize=(16, 5))

    plt.subplot(1, 4, 1)
    plt.imshow(image)
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1, 4, 2)
    plt.imshow(gray, cmap="gray")
    plt.title("Grayscale")
    plt.axis("off")

    plt.subplot(1, 4, 3)
    plt.imshow(mask_raw, cmap="gray")
    plt.title(f"Raw (t={threshold})")
    plt.axis("off")

    plt.subplot(1, 4, 4)
    plt.imshow(mask_cleaned, cmap="gray")
    plt.title("Cleaned")
    plt.axis("off")

    plt.tight_layout()
    plt.show()


# ---- RUN ----

image_path = "data/0_1009_0_0.jpg"

print("Current folder:", os.getcwd())
print("Image exists:", os.path.exists(image_path))

print("Loading image...")
image = load_image(image_path)

print("Converting to grayscale...")
gray = convert_to_grayscale(image)

print("Running custom Otsu...")
t, mask_raw = custom_otsu(gray)

print("Applying morphology...")
mask_cleaned = apply_morphology(mask_raw, 3)

print("Threshold found:", t)

print("Displaying results...")
show_results(image, gray, mask_raw, mask_cleaned, t)

print("Done.")
```

```python
import argparse
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt


# ---- load image ----
def load_image(path):
    # First we check if the file actually exists
    # otherwise it's better to stop early with a clear error
    if not os.path.exists(path):
        raise FileNotFoundError(f"Image not found: {path}")

    # Read image using OpenCV
    img = cv2.imread(path)

    # Sometimes OpenCV fails silently, so we double check
    if img is None:
        raise ValueError("Could not read the image.")

    # Convert from BGR (OpenCV default) to RGB for correct visualization
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


# ---- grayscale ----
def to_gray(img):
    # Thresholding works better on single channel images,
    # so we convert RGB image to grayscale
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)


# ---- custom otsu ----
def custom_otsu(gray):
    # Here we implement a simple version of Otsu thresholding manually
    # instead of using the built-in OpenCV function

    # Compute histogram of pixel intensities
    hist = np.histogram(gray.flatten(), bins=256)[0]
    total = gray.size

    # Compute total sum of intensities (used later)
    sum_total = 0
    for i in range(256):
        sum_total += i * hist[i]

    sumB = 0
    wB = 0
    max_var = 0
    threshold = 0

    # Try every possible threshold value
    for t in range(256):
        wB += hist[t]
        if wB == 0:
            continue

        wF = total - wB
        if wF == 0:
            break

        sumB += t * hist[t]

        # Mean of background and foreground
        mB = sumB / wB
        mF = (sum_total - sumB) / wF

        # This measures how well the two groups are separated
        var_between = wB * wF * (mB - mF) ** 2

        # Keep the threshold that maximizes separation
        if var_between > max_var:
            max_var = var_between
            threshold = t

    # Apply the computed threshold to create binary mask
    _, mask = cv2.threshold(gray, threshold, 255, cv2.THRESH_BINARY)

    return threshold, mask


# ---- morphology ----
def clean_mask(mask, k=3):
    # The raw mask is usually noisy,
    # so we clean it using simple morphological operations

    kernel = np.ones((k, k), np.uint8)

    # Opening removes small noise
    m = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    # Closing fills small holes inside the region
    m = cv2.morphologyEx(m, cv2.MORPH_CLOSE, kernel)

    return m


# ---- keep largest ROI ----
def keep_largest_region(mask):
    # Sometimes multiple regions appear in the mask
    # here we keep only the largest one

    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # If nothing is found, just return original mask
    if len(contours) == 0:
        return mask

    # Select the largest region based on area
    largest = max(contours, key=cv2.contourArea)

    # Create a new empty mask and draw only that region
    new_mask = np.zeros_like(mask)
    cv2.drawContours(new_mask, [largest], -1, 255, thickness=-1)

    return new_mask


# ---- show results ----
def show(img, gray, raw, clean, roi, t):
    # Display all steps side by side to understand the pipeline

    plt.figure(figsize=(16, 5))

    plt.subplot(1, 5, 1)
    plt.imshow(img)
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1, 5, 2)
    plt.imshow(gray, cmap="gray")
    plt.title("Gray")
    plt.axis("off")

    plt.subplot(1, 5, 3)
    plt.imshow(raw, cmap="gray")
    plt.title(f"Raw (t={t})")
    plt.axis("off")

    plt.subplot(1, 5, 4)
    plt.imshow(clean, cmap="gray")
    plt.title("Clean")
    plt.axis("off")

    plt.subplot(1, 5, 5)
    plt.imshow(roi, cmap="gray")
    plt.title("Final ROI")
    plt.axis("off")

    plt.tight_layout()
    plt.show()


# ---- main ----
def main():
    # Define command-line arguments
    parser = argparse.ArgumentParser(
        description="Simple ROI segmentation pipeline"
    )

    parser.add_argument("--image", required=True, help="Path to input image")
    parser.add_argument("--kernel", type=int, default=3, help="Kernel size")
    parser.add_argument("--output", default="outputs", help="Output folder")
    parser.add_argument("--show", action="store_true", help="Show results")

    args = parser.parse_args()

    print("Loading image...")
    img = load_image(args.image)

    # Convert to grayscale
    gray = to_gray(img)

    print("Running custom Otsu...")
    t, raw = custom_otsu(gray)

    print("Cleaning mask...")
    clean = clean_mask(raw, args.kernel)

    print("Extracting ROI...")
    roi = keep_largest_region(clean)

    # Compute a simple metric: how much of the image is ROI
    roi_area = np.sum(roi > 0)
    total_area = roi.size
    roi_percent = (roi_area / total_area) * 100

    print(f"Threshold: {t}")
    print(f"ROI percentage: {roi_percent:.2f}%")

    # Save result
    os.makedirs(args.output, exist_ok=True)
    output_path = os.path.join(args.output, "final_mask.png")
    cv2.imwrite(output_path, roi)

    print(f"Saved to: {output_path}")

    # Optionally display results
    if args.show:
        show(img, gray, raw, clean, roi, t)


if __name__ == "__main__":
    main()
```

```python
import argparse
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt


def load_image(path):
    """
    First step: load the image from disk.

    I added a simple check here because sometimes paths are wrong
    and it's better to fail early with a clear message.
    """
    if not os.path.exists(path):
        raise FileNotFoundError(f"Image not found: {path}")

    img = cv2.imread(path)

    # Sometimes OpenCV fails silently, so it's safer to check this too
    if img is None:
        raise ValueError("Could not read the image file.")

    # OpenCV uses BGR format by default, but matplotlib expects RGB
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


def to_gray(img):
    """
    Convert the image to grayscale.

    Most segmentation techniques (especially thresholding)
    work on single-channel images, so this step is necessary.
    """
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)


def custom_otsu(gray):
    """
    This is a manual implementation of Otsu thresholding.

    Instead of calling OpenCV directly, I compute the threshold
    by analyzing the histogram and maximizing the between-class variance.

    It's not the most optimized implementation, but it's clearer
    and shows how the method actually works.
    """
    hist = np.histogram(gray.flatten(), bins=256)[0]
    total = gray.size

    sum_total = sum(i * hist[i] for i in range(256))

    sumB = 0
    wB = 0
    max_var = 0
    threshold = 0

    # Loop over all possible thresholds
    for t in range(256):
        wB += hist[t]
        if wB == 0:
            continue

        wF = total - wB
        if wF == 0:
            break

        sumB += t * hist[t]

        mB = sumB / wB
        mF = (sum_total - sumB) / wF

        # Between-class variance
        var_between = wB * wF * (mB - mF) ** 2

        if var_between > max_var:
            max_var = var_between
            threshold = t

    # Apply threshold to create binary mask
    _, mask = cv2.threshold(gray, threshold, 255, cv2.THRESH_BINARY)

    return threshold, mask


def clean_mask(mask, k=3):
    """
    After thresholding, the mask is usually noisy.

    Here I use a simple morphological pipeline:
    - Opening removes small noise
    - Closing fills small holes

    The kernel size can be adjusted from CLI.
    """
    kernel = np.ones((k, k), np.uint8)

    m = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    m = cv2.morphologyEx(m, cv2.MORPH_CLOSE, kernel)

    return m


def keep_largest_region(mask):
    """
    In many cases, we are only interested in the main tissue region.

    This function keeps only the largest connected component
    and removes everything else.
    """
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) == 0:
        return mask

    # Select the largest contour based on area
    largest = max(contours, key=cv2.contourArea)

    new_mask = np.zeros_like(mask)
    cv2.drawContours(new_mask, [largest], -1, 255, thickness=-1)

    return new_mask


def show(img, gray, raw, clean, roi, t):
    """
    Just a visualization helper.

    I like to see all intermediate steps together
    to understand what the algorithm is doing.
    """
    plt.figure(figsize=(16, 5))

    plt.subplot(1, 5, 1)
    plt.imshow(img)
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1, 5, 2)
    plt.imshow(gray, cmap="gray")
    plt.title("Gray")
    plt.axis("off")

    plt.subplot(1, 5, 3)
    plt.imshow(raw, cmap="gray")
    plt.title(f"Raw (t={t})")
    plt.axis("off")

    plt.subplot(1, 5, 4)
    plt.imshow(clean, cmap="gray")
    plt.title("Clean")
    plt.axis("off")

    plt.subplot(1, 5, 5)
    plt.imshow(roi, cmap="gray")
    plt.title("Final ROI")
    plt.axis("off")

    plt.tight_layout()
    plt.show()


def main():
    """
    Main pipeline.

    The idea is simple:
    1. Load image
    2. Convert to grayscale
    3. Apply thresholding (custom Otsu)
    4. Clean the mask
    5. Keep only the main region
    6. Save and optionally display results
    """

    parser = argparse.ArgumentParser(description="ROI segmentation pipeline")

    parser.add_argument("--image", required=True, help="Path to input image")
    parser.add_argument("--kernel", type=int, default=3, help="Kernel size")
    parser.add_argument("--output", default="outputs", help="Output folder")
    parser.add_argument("--show", action="store_true", help="Show results")

    args = parser.parse_args()

    print("Loading image...")
    img = load_image(args.image)

    gray = to_gray(img)

    print("Running custom Otsu...")
    t, raw = custom_otsu(gray)

    print("Cleaning mask...")
    clean = clean_mask(raw, args.kernel)

    print("Extracting ROI...")
    roi = keep_largest_region(clean)

    # Some simple statistics, just to have an idea of the segmentation
    roi_area = np.sum(roi > 0)
    total_area = roi.size
    roi_percent = (roi_area / total_area) * 100

    print(f"Threshold: {t}")
    print(f"ROI percentage: {roi_percent:.2f}%")

    os.makedirs(args.output, exist_ok=True)
    output_path = os.path.join(args.output, "final_mask.png")

    cv2.imwrite(output_path, roi)

    print(f"Saved to: {output_path}")

    if args.show:
        show(img, gray, raw, clean, roi, t)


if __name__ == "__main__":
    main()
```
