```python
import numpy as np
from roi_segmentation.main import compute_threshold, apply_threshold


def test_threshold_separates_intensity_groups():
    """
    Here I simulate a simple image with two clear intensity groups.

    The idea is that the threshold should fall somewhere
    between dark and bright values.

    This checks actual behavior, not just execution.
    """

    gray = np.array([
        [10, 20, 15],
        [200, 210, 220]
    ], dtype=np.uint8)

    t = compute_threshold(gray)

    assert 20 < t < 200


def test_apply_threshold_creates_meaningful_segmentation():
    """
    I check that thresholding actually separates dark and bright pixels.

    Dark pixels should become 0
    Bright pixels should become 255
    """

    gray = np.array([
        [10, 200],
        [50, 180]
    ], dtype=np.uint8)

    mask = apply_threshold(gray, 100)

    assert mask[0,0] == 0
    assert mask[0,1] == 255
```
