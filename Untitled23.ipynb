```python
"""
ROI Segmentation Pipeline (Fully Manual Implementation)

In this version, the entire segmentation pipeline is implemented without OpenCV.

The goal is not just to produce a result, but to clearly understand how each
step contributes to the final segmentation.

Each stage is written explicitly:
- image loading (PIL)
- grayscale conversion (manual)
- threshold selection (Otsu-style logic)
- segmentation (binary mask)
- refinement (manual morphology)
- region selection (connected components)
- analysis (ratio + quality)
- visualization (manual overlay + matplotlib)

This makes the pipeline transparent and fully explainable.
"""

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt


# -----------------------------
# Load image
# -----------------------------
def load_image(path):
    """
    The image is loaded using PIL instead of OpenCV.

    The goal is to keep dependencies minimal and focus on understanding
    the data structure (NumPy arrays).

    The image is converted to RGB to ensure consistency.
    """

    img = Image.open(path).convert("RGB")
    return np.array(img)


# -----------------------------
# Convert to grayscale
# -----------------------------
def to_gray(img):
    """
    The segmentation is based on intensity, not color.

    I convert the RGB image into grayscale using a weighted combination
    of channels. These weights reflect how humans perceive brightness.
    """

    r = img[:, :, 0]
    g = img[:, :, 1]
    b = img[:, :, 2]

    gray = 0.299 * r + 0.587 * g + 0.114 * b

    return gray.astype(np.uint8)


# -----------------------------
# Compute threshold (Otsu logic)
# -----------------------------
def compute_threshold(gray):
    """
    The threshold is computed by analyzing intensity distribution.

    The idea is to find a value that best separates foreground
    from background by maximizing variance between classes.
    """

    hist, _ = np.histogram(gray, bins=256, range=(0, 256))

    total = gray.size
    total_sum = np.dot(np.arange(256), hist)

    bg_sum = 0
    bg_weight = 0

    max_var = 0
    best_t = 0

    for t in range(256):
        bg_weight += hist[t]

        if bg_weight == 0:
            continue

        fg_weight = total - bg_weight
        if fg_weight == 0:
            break

        bg_sum += t * hist[t]

        mean_bg = bg_sum / bg_weight
        mean_fg = (total_sum - bg_sum) / fg_weight

        var = bg_weight * fg_weight * (mean_bg - mean_fg) ** 2

        if var > max_var:
            max_var = var
            best_t = t

    return best_t


# -----------------------------
# Apply threshold
# -----------------------------
def apply_threshold(gray, t):
    """
    Each pixel is classified based on the threshold.

    Pixels below threshold → foreground (0)
    Pixels above threshold → background (255)
    """

    mask = np.where(gray < t, 0, 255)
    return mask.astype(np.uint8)


# -----------------------------
# Dilation (manual)
# -----------------------------
def dilation(mask):
    """
    Dilation expands the foreground region.

    If any neighbor is foreground, the pixel becomes foreground.
    """

    h, w = mask.shape
    output = mask.copy()

    for i in range(1, h - 1):
        for j in range(1, w - 1):

            neighborhood = mask[i-1:i+2, j-1:j+2]

            if np.any(neighborhood == 0):
                output[i, j] = 0
            else:
                output[i, j] = 255

    return output


# -----------------------------
# Erosion (manual)
# -----------------------------
def erosion(mask):
    """
    Erosion shrinks the foreground region.

    A pixel stays foreground only if all neighbors are foreground.
    """

    h, w = mask.shape
    output = mask.copy()

    for i in range(1, h - 1):
        for j in range(1, w - 1):

            neighborhood = mask[i-1:i+2, j-1:j+2]

            if np.all(neighborhood == 0):
                output[i, j] = 0
            else:
                output[i, j] = 255

    return output


# -----------------------------
# Refinement (closing)
# -----------------------------
def refine_segmentation(mask):
    """
    This step improves the segmentation.

    Instead of trusting the raw mask, I apply a simple morphological
    closing (dilation + erosion) to smooth boundaries and fill gaps.

    The goal is to make the region more coherent, not perfect.
    """

    return erosion(dilation(mask))


# -----------------------------
# Connected components (manual)
# -----------------------------
def select_main_region(mask):
    """
    The segmentation may contain multiple regions.

    I manually detect connected components using DFS,
    and select the largest one as the main region.
    """

    h, w = mask.shape
    visited = np.zeros_like(mask, dtype=bool)

    best_region = []

    def dfs(x, y):
        stack = [(x, y)]
        region = []

        while stack:
            i, j = stack.pop()

            if i < 0 or i >= h or j < 0 or j >= w:
                continue

            if visited[i, j] or mask[i, j] != 0:
                continue

            visited[i, j] = True
            region.append((i, j))

            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    stack.append((i + dx, j + dy))

        return region

    for i in range(h):
        for j in range(w):
            if not visited[i, j] and mask[i, j] == 0:
                region = dfs(i, j)

                if len(region) > len(best_region):
                    best_region = region

    output = np.full_like(mask, 255)

    for i, j in best_region:
        output[i, j] = 0

    return output


# -----------------------------
# Analysis
# -----------------------------
def analyze(mask):
    """
    This step gives a simple interpretation of the segmentation.

    I compute the ratio of foreground pixels and map it to a label.
    """

    total = mask.size
    roi = np.sum(mask == 0)

    ratio = roi / total

    if ratio > 0.7:
        quality = "High"
    elif ratio > 0.3:
        quality = "Medium"
    else:
        quality = "Low"

    return ratio, quality


# -----------------------------
# Overlay (manual edge)
# -----------------------------
def create_overlay(img, mask):
    """
    Instead of using edge detection libraries, I detect boundaries manually.

    A pixel is considered a boundary if it belongs to the region
    but at least one neighbor is background.
    """

    overlay = img.copy()

    h, w = mask.shape

    for i in range(1, h - 1):
        for j in range(1, w - 1):

            if mask[i, j] == 0:
                neighborhood = mask[i-1:i+2, j-1:j+2]

                if np.any(neighborhood == 255):
                    overlay[i, j] = [255, 0, 0]  # red boundary

    return overlay


# -----------------------------
# Visualization
# -----------------------------
def show_results(img, gray, mask, refined, final, t, ratio, quality):
    """
    This function visualizes the full pipeline.

    It shows each stage and overlays key information
    directly on the final result.
    """

    overlay = create_overlay(img, final)

    plt.figure(figsize=(18, 5))

    titles = ["Original", "Gray", "Mask", "Refined", "Final ROI", "Overlay"]
    images = [img, gray, mask, refined, final, overlay]

    for i in range(6):
        ax = plt.subplot(1, 6, i + 1)

        if i == 0 or i == 5:
            ax.imshow(images[i])
        else:
            ax.imshow(images[i], cmap="gray")

        ax.set_title(titles[i])
        ax.axis("off")

    # overlay text
    ax = plt.subplot(1, 6, 5)

    ax.text(
        0.02, 0.95,
        f"T: {t}\nRatio: {ratio:.2f}\nQuality: {quality}",
        color="red",
        fontsize=10,
        verticalalignment='top',
        transform=ax.transAxes,
        bbox=dict(facecolor="white", alpha=0.85)
    )

    plt.tight_layout()
    plt.show()


# -----------------------------
# Main
# -----------------------------
if __name__ == "__main__":
    import sys

    image_path = sys.argv[1]

    img = load_image(image_path)
    gray = to_gray(img)

    t = compute_threshold(gray)

    mask = apply_threshold(gray, t)
    refined = refine_segmentation(mask)
    final = select_main_region(refined)

    ratio, quality = analyze(final)

    print(f"Threshold: {t}")
    print(f"Ratio: {ratio:.2f}")
    print(f"Quality: {quality}")

    show_results(img, gray, mask, refined, final, t, ratio, quality)
```
