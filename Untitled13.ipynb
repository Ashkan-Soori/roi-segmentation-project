```python
import pytest
from roi_segmentation.main import load_image


def test_load_valid_image():
    """
    This test verifies that a valid image can be loaded successfully.
    """

    img = load_image("data/0_1009_0_0_0.jpg")

    assert img is not None
    assert img.ndim == 3


def test_load_invalid_path():
    """
    This test checks that an error is raised when the image path is invalid.
    """

    with pytest.raises(FileNotFoundError):
        load_image("invalid_path.jpg")
```
