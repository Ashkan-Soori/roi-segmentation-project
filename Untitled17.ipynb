```python
import numpy as np
from roi_segmentation.main import refine_segmentation


def test_refinement_keeps_main_region():
    """
    The main region should remain after refinement.
    """

    mask = np.full((100, 100), 255, dtype=np.uint8)
    mask[30:80, 30:80] = 0

    refined = refine_segmentation(mask)

    assert np.sum(refined == 0) > 0


def test_refinement_reduces_noise():
    """
    Small isolated regions should not grow after refinement.
    """

    mask = np.full((100, 100), 255, dtype=np.uint8)
    mask[0:3, 0:3] = 0

    refined = refine_segmentation(mask)

    assert np.sum(refined == 0) <= np.sum(mask == 0)
```
