```python
import numpy as np
from roi_segmentation.main import compute_threshold, apply_threshold


def test_threshold_range():
    """
    The computed threshold should always be within valid intensity bounds.
    """

    img = np.random.randint(0, 255, (100, 100), dtype=np.uint8)

    t = compute_threshold(img)

    assert 0 <= t <= 255


def test_threshold_binary_output():
    """
    The output mask must be binary (only 0 and 255 values).
    """

    img = np.random.randint(0, 255, (100, 100), dtype=np.uint8)

    mask = apply_threshold(img, 128)

    unique = np.unique(mask)

    assert set(unique).issubset({0, 255})
```
