```python
import numpy as np
from roi_segmentation.main import refine_segmentation, fill_holes


def test_refinement_does_not_expand_noise():
    """
    I create a mask with very small isolated noise.

    After refinement, the segmentation should not grow uncontrollably.

    This checks that the algorithm behaves correctly,
    not just that it runs.
    """

    mask = np.full((50,50), 255, dtype=np.uint8)
    mask[10,10] = 0

    refined = refine_segmentation(mask)

    assert np.sum(refined == 0) <= np.sum(mask == 0)


def test_fill_holes_actually_fills_internal_gaps():
    """
    I create a region with a hole inside.

    A correct hole-filling function should detect that
    the white pixel is surrounded and turn it into foreground.
    """

    mask = np.array([
        [255,255,255],
        [255,0,0],
        [255,255,255]
    ], dtype=np.uint8)

    mask[1,1] = 255  # hole

    result = fill_holes(mask)

    assert result[1,1] == 0
```
